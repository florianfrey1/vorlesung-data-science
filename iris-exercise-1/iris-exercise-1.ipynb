```python
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
import numpy as np
import pandas as pd

# Set default Seaborn style
sns.set()

# Load dat set
iris = pd.read_csv("iris.csv", sep = ';')
```

```python
# get the setosa iris data
setosa = iris[iris.target_names == 'Iris-setosa']
setosa_petal_length = setosa.loc[:, 'petal length (cm)']
setosa_petal_width = setosa.loc[:, 'petal width (cm)']

# get the virginica iris data
virginica = iris[iris.target_names == 'Iris-virginica']
virginica_petal_length = virginica.loc[:, 'petal length (cm)']
virginica_petal_width = virginica.loc[:, 'petal width (cm)']

def scatter_plot(petal_length, petal_width):
    """Print a scatter plot"""
    # Make a scatter plot
    _ = plt.plot(petal_length, petal_width, marker='.', linestyle='none')

    # Label the axes
    _ = plt.xlabel('petal length (cm)')
    _ = plt.ylabel('petal width (cm)')

    # Show the result
    plt.show()

# print scatter plots
scatter_plot(setosa_petal_length, setosa_petal_width)
scatter_plot(virginica_petal_length, virginica_petal_width)

# Compute the covariance matrices
covariance_matrix_setosa = np.cov(setosa_petal_length, setosa_petal_width)
covariance_matrix_virginica = np.cov(virginica_petal_length, virginica_petal_width)

# Print covariance matrices
print(covariance_matrix_setosa)
print(covariance_matrix_virginica)

def plot_covariance_matrix(matrix):
    """Print a covariance matrix heatmap"""
    sns.heatmap(covariance_matrix_setosa, annot=True, fmt='g',
                xticklabels=['petal length', 'petal width'], yticklabels=['petal length', 'petal width'])
    plt.show()

# print covariance matrix heatmaps
plot_covariance_matrix(covariance_matrix_setosa)
plot_covariance_matrix(covariance_matrix_virginica)
```

The covariance matrix already contains in its name its most important property: covariances between variables. In our case, we investigate the covariances between the petal length and the petal width of our irises. Accordingly, we have two variables, which results in a covariance matrix of 2 x 2.

All values on the diagonal of the covariance matrix correspond to the variance of the respective variable, since the covariance with the same variable corresponds to the variance of the variable. Furthermore, the covariance matrix is symmetric, since the order of the variables does not play a role in the covariance (Cov(A,B) = Cov(B,A)).
